```python
import pandas as pd
import numpy as np
```

```python
summaries_train = pd.read_csv("./data/summaries_train.csv")
```

```python
summaries_train[["content", "wording"]].plot(kind="scatter", x="content", y="wording")
#scatterplot

```

```python
angle_deg = 30
angle_rad = np.radians(angle_deg)
rotation_matrix = np.array([[np.cos(angle_rad), -np.sin(angle_rad)],
                            [np.sin(angle_rad), np.cos(angle_rad)]])
summaries_train[['content', 'wording']] = summaries_train[['content', 'wording']].values.dot(rotation_matrix)
```

```python
summaries_train[["content", "wording"]].plot(kind="scatter", x="content", y="wording")
#scatterplot
```
